```python
# 基于卷积神经网络实现图像风格迁移
# 图像风格迁移的算法就是在保证内容图像的内容完整的前提下，将风格图像的风格融入内容图像中
# 定义两个度量值：内容度量值、风格度量值
# 内容度量值用于衡量图片之间的内容差异度程度，风格度量值用于衡量图片之间的风格差异程度
# 对内容图片中的内容和风格图片的风格进行提取，以内容图片为基准将其输入建立的模型中，
# 并不断调整内容度量值和风格度量值，让它们趋近于最小
```

```python
import torch
import torchvision
from torchvision import transforms, models
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt
from torch.autograd import Variable
import copy


# 读取图片，并标准化为张量
transform = transforms.Compose([transforms.Resize([224,224]), 
                                transforms.ToTensor()])
def loadimg(path=None):
    img= Image.open(path)
    img = transform(img)
    img = img.unsqueeze(0)
    return img

# 张量转为图片，并可视化图片
ToPIL = torchvision.transforms.ToPILImage()

def img_show(img, title=None):
    img = img.clone().cpu()
    img = img.view(3, 224,224)
    img = ToPIL(img)
    
    if title is not None:
        plt.title(title)
    plt.imshow(img)

content_img = loadimg("./data/Image/content.jpg")
content_img = Variable(content_img)
style_img = loadimg("./data/Image/style.jpg")
style_img = Variable(style_img)

```

```python
print(content_img.shape,"\n",style_img.shape)
```

```python
plt.figure()
img_show(content_img.data, "Content_Image")
plt.figure()
img_show(style_img.data, "Style_Image")
```

```python
# 图像的内容损失
class Content_loss(torch.nn.Module):
    def __init__(self, weight, target):
        # target是通过卷积获取到的输入图像的内容
        # weight是权重参数，用于控制内容和风格对合成图像的影响程度
        super(Content_loss, self).__init__()
        self.weight = weight
        self.target = target.detach()*weight
        self.loss_fn = torch.nn.MSELoss()
    
    def forward(self, input):
        self.loss = self.loss_fn(input*self.weight, self.target)
        return input
    
    def backward(self):
        self.loss.backward(retain_graph=True)
        return self.loss
```

```python
# 图像的风格损失
# 格拉姆矩阵，n维欧式空间中任意k个向量之间两两的内积所组成的矩阵，
# 称为这k个向量的格拉姆矩阵(Gram matrix)，很明显，这是一个对称矩阵
# 格拉姆矩阵可以度量各个维度自己的特性以及各个维度之间的关系
class Gram_matrix(torch.nn.Module):
    def forward(self, input):
        a, b ,c , d = input.size()
        feature = input.view(a*b, c*d)
        gram = torch.mm(feature, feature.t())
        return gram.div(a*b*c*d)
        
class Style_loss(torch.nn.Module):
    def __init__(self, weight, target):
        super(Style_loss, self).__init__()
        self.weight = weight
        self.target = target.detach()*weight
        self.loss_fn = torch.nn.MSELoss()
        self.gram = Gram_matrix()
        
    def forward(self, input):
        self.Gram = self.gram(input.clone())
        self.Gram.mul_(self.weight)
        self.loss = self.loss_fn(self.Gram, self.target)
        return input
    
    def backward(self):
        self.loss.backward(retain_graph=True)
        return self.loss
```

```python
# 迁移一个卷积圣经网络的特征提取部分，即卷积相关的部分
cnn = models.vgg16(pretrained=True).features

# 设置在卷积过程中的哪一层提取内容、风格
content_layer = ["Conv_3"]
style_layer = ["Conv_1", "Conv_2", "Conv_3", "Conv_4", ""]

content_losses = []
style_losses = []

# 设置影响权重
content_weight = 1
style_weight = 100000

# 自定义模型
new_model = torch.nn.Sequential()
model = copy.deepcopy(cnn)
gram = Gram_matrix()

Use_gpu = torch.cuda.is_available()
if Use_gpu:
    new_model = new_model.cuda()
    gram = gram.cuda()
    
index = 1
for layer in list(model)[:8]: # 截取使用迁移模型特征提取部分的前8层
    if isinstance(layer, torch.nn.Conv2d):
        name = "Conv_" + str(index)
        new_model.add_module(name, layer) # 向新模型中添加指定的层次模块
        if name in content_layer:
            target = new_model(content_img).clone()
            content_loss = Content_loss(content_weight, target)
            new_model.add_module("Content_loss_" + str(index), content_loss)
            content_losses.append(content_loss)
        
        if name in style_layer:
            target = new_model(style_img).clone()
            target = gram(target)
            style_loss = Style_loss(style_weight, target)
            new_model.add_module("Style_los_" + str(index), style_loss)
            style_losses.append(style_loss)
        
    if isinstance(layer, torch.nn.ReLU):
        new_model.add_module("ReLu_" + str(index), layer)
        index += 1
    
    if isinstance(layer, torch.nn.MaxPool2d):
        new_model.add_module("MaxPool_" + str(index), layer)

print(new_model)
```

```python
# 训练模型
input_img = content_img.clone()
parameter = torch.nn.Parameter(input_img.data)
optimizer = torch.optim.LBFGS([parameter])

epoch_n = 300
epoch = [0]

while epoch[0] <= epoch_n:
    def closure():
        optimizer.zero_grad()
        style_score = 0
        content_score = 0
        parameter.data.clamp_(0,1)
        new_model(parameter)
        for sl in style_losses:
            style_score += sl.backward()
        for cl in content_losses:
            content_score += cl.backward()
            
        epoch[0] += 1
        if epoch[0] % 50 == 0:
            print("Epoch:{} Style Loss:{:.4f} Content Loss:{:.4f}".format(
                    epoch[0], style_score.data, content_score.data))
        return style_score + content_score
    optimizer.step(closure)
        
```

```python
# 可是最终结果
parameter.data.clamp_(0,1)
plt.figure()
img_show(parameter.data, "Output Image")
```
